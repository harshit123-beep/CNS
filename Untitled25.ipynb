```python
from scapy.all import IP, TCP, UDP
import numpy as np

# Dummy ML model (replace with your trained model)
class DummyModel:
    def predict(self, X):
        # Simple logic: if packet length > 60 → intrusion
        return [1 if x[0] > 60 else 0 for x in X]

model = DummyModel()

# Packet processing function
def process_packet(packet):
    if packet.haslayer(IP):
        src = packet[IP].src
        dst = packet[IP].dst
        proto = packet[IP].proto

        # Feature extraction
        length = len(packet)
        features = [[length, proto]]

        # Prediction
        prediction = model.predict(features)

        print(f"\n📦 Packet Flow:")
        print(f"Source: {src} → Destination: {dst}")
        print(f"Protocol: {proto} | Length: {length}")

        if prediction[0] == 0:
            print("Result: Normal Traffic ✅")
        else:
            print("Result: Intrusion Detected 🚨")

# Create sample packets
pkt1 = IP(src="192.168.1.1", dst="192.168.1.2")/TCP()
pkt2 = IP(src="10.0.0.5", dst="8.8.8.8")/UDP()

# Simulate packet flow
process_packet(pkt1)
process_packet(pkt2)
```
